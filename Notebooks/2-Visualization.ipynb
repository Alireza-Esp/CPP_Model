### ✅ Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

### ✅ Getting raw data from github repository

```python
df=pd.read_csv("https://raw.githubusercontent.com/Alireza-Esp/CPP_Model/main/Data/CarsData.csv")
```

```python
df.head(20)
```

### ✅ Visualization

```python
sns.countplot(data=df,x="Manufacturer",palette="CMRmap")
plt.xlabel("name of company",fontsize=10,color='black')
plt.ylabel("count",fontsize=10,color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df,x="year",palette='icefire')
plt.xlabel("year",fontsize=10,color='black')
plt.ylabel("count",fontsize=10,color='black')
plt.xticks(rotation=90)
plt.show()
```

```python
labels=df['fuelType'].value_counts().index
sizes=df['fuelType'].value_counts()
colors=['red','pink','blue','black','yellow','orange','purple']
plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,colors=colors,startangle=0,)
plt.show()
```
