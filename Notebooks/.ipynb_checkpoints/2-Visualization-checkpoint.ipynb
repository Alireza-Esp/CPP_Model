### ✅ Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

### ✅ Getting raw data from github repository

```python
Raw_Data_Url = "https://raw.githubusercontent.com/Alireza-Esp/CPP_Model/main/Data/CarsData.csv"
```

```python
Raw_Data = pd.read_csv(Raw_Data_Url)
```

```python
Raw_Data.head(20)
```

### ✅ Visualization

#### 🔹 Comparing manufacturers respect to the count of cars (in dataset)

```python
sns.countplot(data=Raw_Data, x="Manufacturer", palette="CMRmap")
plt.xlabel("Companies", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing distribution of cars data across the years

```python
sns.countplot(data=Raw_Data, x="year" ,palette='icefire')
plt.xlabel("Year", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=90)
plt.show()
```

#### 🔹 Showing distribution of cars price

```python
sns.displot(data=Raw_Data, y="price", bins=100, color="purple")
plt.xlabel("Count", fontsize=12, color='black')
plt.ylabel("Price($)", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing percentages of cars gearbox type

```python
plt.figure(figsize=(7,7))
plt.pie(Raw_Data['transmission'].value_counts(),
        labels=Raw_Data['transmission'].value_counts().index,
        startangle=0,
        autopct='%1.f%%',
        colors=sns.color_palette(palette="Set2"))
plt.legend()
plt.show()
```

#### 🔹 Showing distribution of miles used for cars

```python
sns.displot(data=Raw_Data, x="mileage", color="orange")
plt.xlabel("Miles Used", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing distribution of mile pre gallon for all cars

```python
sns.displot(data=Raw_Data, x="mpg", color="brown", bins=100)
plt.xlabel("Mile per Gallon", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

#### 🔹 Showing distribution of engines size

```python
sns.displot(data=Raw_Data, y="engineSize", color="teal", bins=30)
plt.xlabel("Mile per Gallon", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing percentages of cars fuel type

```python
plt.figure(figsize=(7,7))
plt.pie(Raw_Data['fuelType'].value_counts(),
        labels=Raw_Data['fuelType'].value_counts().index,
        startangle=0,
        autopct='%1.f%%',
        colors=sns.color_palette(palette="Set3"),
        explode=[0, 0, 0.5, 1.4, 0])
plt.legend()
plt.show()
```

#### 🔹 Showing distribution of price respect to year and gearbox type, mileage, fuel type, engine size

```python
for col in ["transmission", "mileage", "fuelType", "engineSize"]:
    sns.scatterplot(Raw_Data, x="year", y="price", hue=col , marker="s", s=50)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Price", fontsize=12)
    plt.show()
```

#### 🔹 Separating the manufacturers and types of fuels

```python
sns.countplot(data=Raw_Data, x="Manufacturer", hue="fuelType",
              palette=["#e50000", "#448ee4", "#ff7fa7", "#c6f808", "#1fb57a"])
plt.xlabel("Companies", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing distribution of manufacturers and gearboxes types

```python
sns.countplot(data=Raw_Data, x="Manufacturer", hue="transmission",
              palette=["#f5054f", "#6832e3", "#04f489", "#d725de"])
plt.xlabel("Companies", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing distribution of manufacturers according to tax

```python
sns.boxplot(data=Raw_Data, x="Manufacturer", y="tax")
plt.xlabel("Manufacturer", fontsize=12)
plt.ylabel("Tax", fontsize=12)
plt.xticks(rotation=30)
plt.show()
```

#### 🔹 Showing distribution of manufacturers and engines size

```python
Raw_Data["Engine Size(lit)"] = pd.cut(Raw_Data["engineSize"],
                                      bins=[0, 2, 4, 7],
                                      labels=["Medium: 0 < ES <= 2", "High: 2 < ES <= 4", "Very High: 4 < ES"])
sns.countplot(data=Raw_Data, x="Manufacturer", hue="Engine Size(lit)",
              palette=["#f8481c", "#04d9ff", "#61e160"])
plt.xlabel("Companies", fontsize=12, color='black')
plt.ylabel("Count", fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
print(Raw_Data[Raw_Data["Engine Size(lit)"] == "Very High: 4 < ES"]["Manufacturer"].value_counts().to_dict())
Raw_Data.drop(columns=["Engine Size(lit)"], inplace=True)
```

#### 🔹 Showing distribution of manufacturers and tax

```python
Raw_Data['Tax($)'] = pd.cut(Raw_Data['tax'], bins=[0,150,300,450,600],
                           labels=["Low: 0<tax<150",
                                   "Medium: 150<tax<300",
                                   "High: 300<tax<450",
                                   "Very High: 450<tax<600"])
sns.countplot(data=Raw_Data, x="Manufacturer", hue='Tax($)', palette=['#f075e6', '#c6f808', '#cf0234', '#7bc8f6'])
plt.xlabel('Companies', fontsize=12, color='black')
plt.ylabel('Counts', fontsize=12, color='black')
plt.xticks(rotation=30)
plt.show()
print(Raw_Data[Raw_Data['Tax($)'] == 'High: 300<tax<450']['Manufacturer'].value_counts().to_dict())
print(Raw_Data[Raw_Data['Tax($)'] == 'Very High: 450<tax<600']['Manufacturer'].value_counts().to_dict())
Raw_Data.drop(columns=["Tax($)"], inplace=True)
```
